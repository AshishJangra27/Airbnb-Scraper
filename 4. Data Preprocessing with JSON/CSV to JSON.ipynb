```python
import json
import pandas as pd

df = pd.read_csv('airbnb_v2.csv')
df.head()
```

### 1. Deleting the Columns that are not needed

```python
del df['Unnamed: 0']
del df['features']
del df['safety_rules']
del df['hourse_rules']
del df['checkin']
del df['checkout']

df.head()
```

### 2. Prepairing Dictionary

```python
data = df.values

dct = {}

for i in data:
    
    id_       = i[0] #
    name      = i[1] #
    rating    = i[2] #
    review    = i[3] #
    host_name = i[4] #
    host_id   = i[5] #
    address   = i[6] #
    amenities = i[7] #
    img_links = i[8]
    price     = i[9]  #
    country   = i[10] #
    bathrooms = i[11] #
    beds      = i[12] #
    guests    = i[13] #
    toiles    = i[14] #
    bedrooms  = i[15] #
    studio    = i[16] #
    
    dct[id_] = {'name' : name, 'rating' : rating, 'review' : review, 'host_name' : host_name, 'host_id' : host_id,
                'address' : address , 'price' : price, 'country' : country, 'bathrooms' : bathrooms,
                'beds' : beds, 'guests' : guests, 'toilets' : toiles, 'bedrooms' : bedrooms, 'studios' : studio,
                'amenities' : amenities.split(','), 'img_links' : img_links.split(' ')}
```

### 3. Saving JSON

```python
js = json.dumps(dct)

fd = open('airbnb.json','w')
fd.write(js)
fd.close()
```
